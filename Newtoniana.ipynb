```python
%%manim -qm ThreeDCameraRotation

from manim import *

class ThreeDCameraRotation(ThreeDScene):
    def construct(self):
        self.camera.background_color = BLACK

        plane = NumberPlane(x_range=(-5, 5), y_range=(-5, 5))
        self.add(plane)

        self.set_camera_orientation(phi=74 * DEGREES, theta=-45 * DEGREES)

        axes = ThreeDAxes(tips=False)
        max_radius = 4
        min_radius = 1
        circle = Circle(radius=max_radius, color=WHITE)

        self.add(axes)

        sphere = Sphere(radius=0.1, color=RED).shift(1.5 * RIGHT)
        connecting_line = Line(ORIGIN, sphere.get_center(), color=RED)
        self.add(sphere, connecting_line, circle)

        # Modificar la altura y el radio del cilindro
        cylinder = Cylinder(height=2, radius=0.1, color=BLUE_D)
        self.add(cylinder)

        # Ajustar la posicin del punto superior del cilindro
        top_point_cylinder = cylinder.get_center() + 0.5 * cylinder.get_height() * normalize(cylinder.get_height())
        top_point_cylinder[2] = 0  # Establecer el componente z a cero
        connecting_line_cylinder = Line(
            top_point_cylinder, np.array([0, 0, 1]), color=GREEN
        )
        self.add(connecting_line_cylinder)

        def update_circle(circle, alpha):
            new_radius = max(min_radius, max_radius - max_radius * alpha)
            circle.become(Circle(radius=new_radius, color=WHITE))
            connecting_line.put_start_and_end_on(ORIGIN, sphere.get_center())

            # Ajustar la posicin del punto superior del cilindro
            top_point_cylinder = cylinder.get_center() + 0.5 * cylinder.get_height() * normalize(cylinder.get_height())
            top_point_cylinder[2] = 0  # Establecer el componente z a cero
            connecting_line_cylinder.put_start_and_end_on(
                sphere.get_center(), np.array([0, 0, 1])
            )

        self.play(Write(circle))
        self.play(
            MoveAlongPath(sphere, circle),
            UpdateFromAlphaFunc(circle, update_circle),
            run_time=120,
            rate_func=linear,
        )

        self.wait()
```
